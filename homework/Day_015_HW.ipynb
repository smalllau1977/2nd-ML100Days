```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
"""
Your Code Here
"""
#建立一個 10 x 10, 數值分布自 -1.0 ~ 1.0 的矩陣
matrix = np.random.uniform(-1,1,(10,10)) 
matrix
```

```python
#建立一個 1000 x 3, 數值分布為 -1.0 ~ 1.0 的矩陣
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = np.random.uniform(-1,1,(nrow,ncol)) 
matrix
```

```python
# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
"""
Your Code Here
Please replace "..." to correct plot function
"""
grid.map_upper(plt.scatter, alpha = 0.2)
grid.map_diag(sns.kdeplot )
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.show()
```

```python
#請用 numpy 建立一個 1000 x 3, 數值分布為常態分佈的矩陣
nrow = 1000
ncol = 3
"""
Your Code Here
"""
mu = 1
sigma = 0.1
matrix = np.random.normal(loc=mu, scale=sigma, size=[nrow,ncol])

matrix
```

```python
# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice).reset_index()

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
"""
Your Code Here
Please replace "..." to correct plot function
"""
grid.map_upper(plt.scatter, alpha = 0.2)
grid.map_diag(sns.kdeplot )
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.show()
```
